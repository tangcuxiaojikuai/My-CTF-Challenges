```python
from Crypto.Util.number import *

dic = ['a@4A', 'b8B', 'cC', 'dD', 'e3E', 'fF', 'g9G', 'hH', 'iI', 'jJ', 'kK', 'l1L', 'mM', 'nN', 'o0O', 'pP', 'qQ', 'rR', 's$5S', 't7T', 'uU', 'vV', 'wW', 'xX', 'yY', 'z2Z']
c1 = "nSsCtf{H@pPy_A7h_4nNiv3R$arY_ns$C7F_@nD_l3t$_d0_7h3_LllE3t$pEAk!}"
c2 = 9114319649335272435156391225321827082199770146054215376559826851511551461403
x = 2**255+95
n = 2^256

t = []
d = 0
for i in c1[::-1]:
    if(i == "}" or i == "!" or i == "_"):
        c2 -= ord(i)*pow(x, d, n)
        c2 = c2 % n
    elif(i == "{"):
        for j in "NSSCTF{"[::-1]:
            c2 -= ord(j)*pow(x, d, n)
            c2 = c2 % n
            d += 1
        break
    else:
        for j in dic:
            if(i in j):
                c2 -= ord(j[0])*pow(x, d, n)
                c2 = c2 % n
                for _ in j[1:]:
                    t.append((ord(_)-ord(j[0]))*pow(x, d, n) % n)
                break
    d += 1

L = block_matrix(ZZ, [
    [2*identity_matrix(len(t)), 0, n*Matrix(ZZ, t).T],
    [Matrix(ZZ, [-1]*len(t)), 1, n*Matrix(ZZ, [-c2])],
    [0, 0, n^2]
])
L = L.BKZ(block_size=20)

for i in L:
    print(i[:-2][::-1])
```
